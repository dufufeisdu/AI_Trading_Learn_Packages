```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ts = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2000', periods=1000))
ts = ts.cumsum()
```

```python
ts.plot()
```

```python
df = pd.DataFrame(np.random.randn(1000, 4), index=ts.index, columns=list('ABCD'))
df  = df.cumsum()
plt.figure()
df.plot()
```

```python
df.plot(x=ts.index,y='A')
```

```python
df.iloc[:10,:].plot(x=ts.index[:10],y='A',kind='box')
```

```python
df.iloc[:10,:].plot(x=ts.index[:10],y='A',kind='density')
```

```python
df.iloc[:10,:].plot(x=ts.index[:10],y='A',kind='kde')
```

```python
df.iloc[:10,:].plot(x=ts.index[:10],y='A',kind='hist')
```

```python
dd = df.iloc[:10,:]
dd.plot.bar()
```

```python
dd['A'].diff().plot.bar()
```

```python
df = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])
df['b'] = df['b'] + np.arange(1000)
df.plot.hexbin(x='a', y='b', gridsize=25)
```

```python
series = pd.Series(3 * np.random.rand(4), index=['a', 'b', 'c', 'd'], name='series')
series.plot.pie(figsize=(6,6))
```
