#### <center><span style="color:red">Use pd.IndexSlice to silice multi index<span></center>

```python
def createI(prefix,n):
    return [prefix+str(i)  for i in range(n) ]
```

```python
iIndex=pd.MultiIndex.from_product([createI('A',4),
                                    createI('B',2),
                                     createI('C',4),
                                      createI('D',2)])
```
