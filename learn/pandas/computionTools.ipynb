## Rolling vs expanding: rolling is based on the first data of window, while expanding is based on the first data of the whole set. Expanding don't have the window parameter

```python
dfb.rolling(window=len(dfb), min_periods=1).mean()[:5]==dfb.expanding(min_periods=1).mean()[:5]
```

```python
sn.expanding().sum()==sn.cumsum().fillna(method='ffill')
```

## Exponentially Weighted Window: one must be set: span, center of mass, half-life and alpha to the EW functions
* This is a way to adjust data based on the past data: [x0,x1,x2]=>[y0,y1,y2]
* $\alpha$ is very important and related to one of the three: span, center of mass, half-life
* When $\alpha$ is 1 it will become [x0,x1,x2]=>[x0,x1,x2]
